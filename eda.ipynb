Agora irei tratar as colunas que possuem valores em strings.

```python
print(df.iloc[3])
```

```python
def tratar_strings(df):
    import re

    # Remove "R$" e converte para float nas colunas de valores monetários
    for col in ['valor', 'condominio', 'iptu']:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(r'[R$\s.]', '', regex=True)
            .str.replace(',', '.')
            .replace('nan', None)
            .astype(float)
        )

    # Separa endereço em rua, bairro e número
    endereco_split = df['endereco'].str.split(' - ', expand=True)
    if endereco_split.shape[1] >= 3:
        df['rua'] = endereco_split[0]
        df['bairro'] = endereco_split[1]
        df['numero'] = endereco_split[2]
    else:
        df['rua'] = endereco_split[0]
        df['bairro'] = endereco_split[1] if endereco_split.shape[1] > 1 else None
        df['numero'] = None

    # Extrai número de metros quadrados de 'floorSize'
    df['floorSize'] = (
        df['floorSize']
        .astype(str)
        .str.extract(r'(\d+)', expand=False)
        .replace('nan', None)
        .astype(float)
    )

    # Extrai número de quartos e banheiros
    for col in ['numberOfRooms', 'numberOfBathroomsTotal']:
        df[col] = (
            df[col]
            .astype(str)
            .str.extract(r'(\d+)', expand=False)
            .replace('nan', None)
            .astype(float)
        )

    return df
```

```python
tratar_strings(df)
```
